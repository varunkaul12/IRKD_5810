```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
# Since the dataset doesn't contain the column header, we need to specify it manually.   
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'annual-income']  
 
# Importing dataset    
data = pd.read_csv("C:\\Users\\T0253VK\\Desktop\\HOME\\ForMS\\Information_Retreival_5810\\Assignment\\Census+Income\\adult.data",names = col_names)
print(data.shape)
print(col_names)
data.head()
```

```python
#Get the data information
data.info()
```

```python
#Create a dataframe and pring the values 
df = pd.DataFrame(data)
print(df.describe())
print((pd.DataFrame(data,columns=['annual-income'])).describe())
```

```python
# For this plot, we will be using the seaborn library as it provides more flexibility with dataframes.   
sns.barplot(data.gender, data['hours-per-week'], hue=data['annual-income'])  
plt.show()
```

#### we can infer that males and females with annual income less than 50K tend to work more per week.

```python
# Using Countplot to count number of males and females in the dataset.  
sns.countplot(data.gender)  
plt.show()  
```

#### we can infer that there are more number of males than females in the dataset

```python
# ‘hue’ is used to visualize the effect of an additional variable to the current distribution.  
sns.countplot(data.gender, hue=data['annual-income'])  
plt.show() 
```

#### we can infer that out of 32,500 (approx) people, only 8000 people have income greater than 50K, out of which only 1000 of them are females

```python
# Creating a box plot  
fig, ax = plt.subplots(figsize=(15, 8))  
sns.boxplot(x='relationship', y='hours-per-week', hue='annual-income', data=data, ax=ax)  
ax.set_title('Annual Income of people based on relationship and hours-per-week')  
plt.show()  
```

#### People with the same relationship status and an annual income more than 50K often work for more hours per week. 
#### People who have a child and earn less than 50K tend to have more flexible working hours.
